```python
# Q1. Creating a DataFrame
# Create a Pandas DataFrame using a Python dictionary with the following columns:Product, Region, Sales, Quantity

# Creating a dictionary
data = {
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Keyboard', 'Mouse'],
    'Region': ['East', 'West', 'North', 'South', 'East'],
    'Sales': [50000, 30000, 20000, 5000, 3000],
    'Quantity': [5, 10, 7, 15, 20]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print(df)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("C:\\Users\\Renu Sharma\\Downloads\\Retail_dataset.csv")
```

```python
# Q2. Data Inspection

# Display first 5 rows
print("First 5 Rows:")
print(df.head())

# Display last 5 rows
print("\nLast 5 Rows:")
print(df.tail())

# Display dataset structure
print("\nDataset Information:")
print(df.info())

# Display shape of dataset
print("\nShape of Dataset:")
print(df.shape)
```

```python
# Q3. Column Selection

# Select required columns
selected_df = df[['Product', 'Region', 'Sales']]

# Display result
print(selected_df)
```

```python
# Q4. Indexing using .loc

# Display row with index 10
print("Row with index 10:")
print(df.loc[10])

# Display Sales value of row with index 20
print("\nSales value at index 20:")
print(df.loc[20, 'Sales'])
```

```python
# Q5. Indexing using .iloc
# Using .iloc, display: 

# 1. First 6 rows and first 4 columns
print("First 6 rows and first 4 columns:")
print(df.iloc[:6, :4])

# 2. Sales value from 3rd row
print("\nSales value from 3rd row:")
print(df.iloc[2, 2])   # row index 2 = 3rd row, column index 2 = Sales
```

```python
# Q6. Row Slicing

# 1. First 10 rows
print("First 10 rows:")
print(df.head(10))

# 2. Rows 15 to 25
print("\nRows 15 to 25:")
print(df.iloc[15:26])   # 26 is excluded, so includes up to 25
```

```python
# Q7. Filtering Data

# 1. Rows where Region = East
print("Rows where Region = East:")
print(df[df['Region'] == 'East'])

# 2. Rows where Sales > 1500
print("\nRows where Sales > 1500:")
print(df[df['Sales'] > 1500])
```

```python
# Q8. Multiple Condition Filtering

# Filter: Region is West AND Sales > 1200
filtered_df = df[(df['Region'] == 'West') & (df['Sales'] > 1200)]

# Display result
print(filtered_df)
```

```python
# Q9. Descriptive Statistics

# Generate summary statistics
desc = df.describe()

print("Descriptive Statistics:")
print(desc)

# Required values
mean_sales = df['Sales'].mean()
max_quantity = df['Quantity'].max()
min_sales = df['Sales'].min()

print("\nMean Sales Value:", mean_sales)
print("Maximum Quantity Sold:", max_quantity)
print("Minimum Sales Value:", min_sales)
```

```python
# Q10. Aggregation Analysis

# Calculations using Sales column
total_sales = df['Sales'].sum()
average_sales = df['Sales'].mean()
max_sales = df['Sales'].max()
min_sales = df['Sales'].min()

print("Total Sales:", total_sales)
print("Average Sales:", average_sales)
print("Maximum Sales:", max_sales)
print("Minimum Sales:", min_sales)
```
