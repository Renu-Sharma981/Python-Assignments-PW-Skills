```python
import pandas as pd
df = pd.read_csv("C:\\Users\\Renu Sharma\\Downloads\\retail_dataset .csv")
df.head()
```

```python
import numpy as np
```

```python
# 1.display the first 5 rows along with the dataset information to understand its structure.
print(df.head())
print(df.info())    
```

```python
# 2. Inspect the dataset and identify missing values in each column.
# Display the total number of missing values present in the dataset.
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
total_missing = missing_values.sum()
print(f"Total number of missing values in the dataset: {total_missing}")
```

```python
# 3. Replace missing values in the Quantity and Price columns 
# with their respective mean values using Pandas.
# 3. Replace missing values in Quantity and Price columns with their mean
# Copy-on-Write friendly method

df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())


print("Missing values after filling:")
print(df[['Quantity', 'Price']].isnull().sum())
```

```python
# 4. Remove all rows where the Product Category or Region column contains missing values.
df.dropna(subset=['Product Category', 'Region'], inplace=True)
print("Dataset after removing rows with missing values in Product Category or Region:")
print(df.head())
```

```python
# 5.Create a new column called Revenue using NumPy, where:
# Revenue = Quantity × Price
#Display the updated dataset with the new column
df['Revenue'] = df['Quantity'] * df['Price']
print("Dataset with the new Revenue column:")
print(df.head())
```

```python
# 6.Using NumPy, calculate the total revenue generated from all transactions in the dataset.
total_revenue = df['Revenue'].sum()
print(f"Total revenue generated from all transactions: {total_revenue}")
```

```python
# 7. Group the dataset by Product Category and calculate the total revenue generated by each category.
revenue_by_category = df.groupby('Product Category')['Revenue'].sum()
print("Total revenue generated by each Product Category:")
print(revenue_by_category)
```

```python
# 8.Based on the revenue calculated for each product category, identify the top 3 categories with the
#highest revenue and the bottom 3 categories with the lowest revenue.


sorted_revenue = revenue_by_category.sort_values(ascending=False)

# Top 3 Categories
top_3_categories = sorted_revenue.head(3)

# Bottom 3 Categories
bottom_3_categories = sorted_revenue.tail(3)

print("--- Top 3 Product Categories ---")
print(top_3_categories)

print("\n--- Bottom 3 Product Categories ---")
print(bottom_3_categories)

```

```python
# 9. Group the dataset bmmy Region and calculate the total revenue generated in each region. 
# Identify the region with the highest revenue and the region with the lowest revenue.
revenue_by_region = df.groupby('Region')['Revenue'].sum()
print("Total revenue generated by each Region:")    
print(revenue_by_region)
highest_revenue_region = revenue_by_region.idxmax()
lowest_revenue_region = revenue_by_region.idxmin()
print(f"Region with the highest revenue: {highest_revenue_region}")
print(f"Region with the lowest revenue: {lowest_revenue_region}")
```

```python
# 10. Extract the month from the Date column using Pandas and group the dataset by month 
# to calculate the total monthly revenue. Also, calculate the mean, median, and standard deviation 
# of the Revenue column using NumPy.
# 10. Extract month and calculate stats
import pandas as pd
import numpy as np

df = pd.read_csv('retail_dataset .csv') 
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Month'] = df['Date'].dt.month
df['Revenue'] = df['Price'] * df['Quantity']

# Monthly Revenue
monthly_revenue = df.groupby('Month')['Revenue'].sum()
print("Total Monthly Revenue:")
print(monthly_revenue)

rev_mean = np.mean(df['Revenue'])
rev_median = np.median(df['Revenue'])
rev_std = np.std(df['Revenue'])

print(f"\nMean Revenue: {rev_mean}")
print(f"Median Revenue: {rev_median}")
print(f"Standard Deviation: {rev_std}")
```
