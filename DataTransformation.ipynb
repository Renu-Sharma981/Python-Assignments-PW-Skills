```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv(r"C:\Users\Renu Sharma\Downloads\employee_productivity_dataset.csv")
```

```python
# Q1. Handle Missing Values (Basic Cleaning)

# Fill missing values
df["Age"].fillna(df["Age"].median(), inplace=True)

df["Salary"].fillna(df["Salary"].mean(), inplace=True)

df["Hours_Worked_Per_Week"].fillna(
    df["Hours_Worked_Per_Week"].median(),
    inplace=True
)

df["Performance_Score"].fillna(
    df["Performance_Score"].mean(),
    inplace=True
)

# Display updated dataset
print(df)
```

```python
# Q2. Label Encoding

from sklearn.preprocessing import LabelEncoder

# Create LabelEncoder object
le = LabelEncoder()

# Apply Label Encoding
df["Gender"] = le.fit_transform(df["Gender"])

df["Department"] = le.fit_transform(df["Department"])

# Display updated columns
print(df[["Gender", "Department"]])
```

```python
# Q3. One-Hot Encoding

# Apply One-Hot Encoding
df_encoded = pd.get_dummies(
    df,
    columns=["Work_Mode", "Location"]
)

# Display updated dataset
print(df_encoded)

# Check number of new columns created
print("Total Columns After Encoding:", len(df_encoded.columns))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Q4. Normalization (Min-Max Scaling)

# Create scaler object
scaler = MinMaxScaler()

# Normalize selected columns
df[["Salary", "Hours_Worked_Per_Week"]] = scaler.fit_transform(
    df[["Salary", "Hours_Worked_Per_Week"]]
)

# Display results
print(df[["Salary", "Hours_Worked_Per_Week"]])
```

```python
# Q5. Standardization (Scaling)

from sklearn.preprocessing import StandardScaler

# Create StandardScaler object
scaler = StandardScaler()

# Apply Standardization
df[["Age", "Projects_Completed"]] = scaler.fit_transform(
    df[["Age", "Projects_Completed"]])

# Display transformed values
print(df[["Age", "Projects_Completed"]])
```

```python
# Q6. Compare Scaling Methods

from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Create scaler objects
minmax = MinMaxScaler()
standard = StandardScaler()

# Apply scaling on Salary column
df["Salary_MinMax"] = minmax.fit_transform(df[["Salary"]])

df["Salary_Standard"] = standard.fit_transform(df[["Salary"]])

# Display both results side by side
print(df[["Salary", "Salary_MinMax", "Salary_Standard"]])
```

```python
# Q7. Build Preprocessing Pipeline

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import (
    OneHotEncoder,
    StandardScaler
)

# Define categorical and numerical columns
categorical_cols = [
    "Gender",
    "Department",
    "Work_Mode",
    "Location"
]

numerical_cols = [
    "Age",
    "Salary",
    "Hours_Worked_Per_Week",
    "Performance_Score",
    "Projects_Completed"
]

# Create preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        (
            "num",
            StandardScaler(),
            numerical_cols
        ),
        (
            "cat",
            OneHotEncoder(),
            categorical_cols
        )
    ]
)

# Create pipeline
pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor)
    ]
)

# Display pipeline
print(pipeline)
```

```python
# Q8. Apply Pipeline

# Apply pipeline on dataset
transformed_data = pipeline.fit_transform(df)

# Display transformed dataset
print(transformed_data)

# Display shape of final dataset
print("Shape of Final Dataset:", transformed_data.shape)
```
