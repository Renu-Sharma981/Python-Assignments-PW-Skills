```python
# 5. Write a Python program to check whether a number is even or odd using conditional
#statements.
num =int(input("Enter a number"))
if num % 2 == 0:
    print(num, "is an even number")
else:
    print(num, "is an odd number")
```

```python
# 7. Create a list of five numbers and print the maximum number from the list.

numbers = [10, 20, 5, 15, 30]
max_number = max(numbers)
print("The maximum number is:", max_number)
```

```python
# 8. Write a Python program to remove duplicate values from a list using a set.

my_list = [1, 2, 3, 2, 4, 1, 5]
unique_values = set(my_list)
print("Unique values:", unique_values)
```

```python
# 9.  Write a function that returns the square of a number.

def square(num):
    return num ** 2    

# Example usage:
result = square(5)
print("The square of 5 is:", result)
```

```python
# 10. Write a Python program to count how many times a number appears in a list.
# Example List
# [2,3,4,2,5,2]

numbers = [2, 3, 4, 2, 5, 2]
number_to_count = 2
count = numbers.count(number_to_count)
print("The number 2 appears", count, "times")
```
