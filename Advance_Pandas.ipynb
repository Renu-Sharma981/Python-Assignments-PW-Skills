```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv(r"C:\Users\Renu Sharma\Downloads\Advanced_pandas_sales_dataset.csv")

# Q1 Dataset Exploration

#  Display first 10 rows
print("First 10 rows:")
print(df.head(10))

#  Display dataset structure
print("\nDataset Info:")
print(df.info())

#  Number of rows and columns
print("\nShape of dataset (rows, columns):")
print(df.shape)

#  Data types of each column
print("\nData types of each column:")
print(df.dtypes)
```

```python
# Q2 Sales Filtering Analysis

# Orders where Sales > 3500
high_sales_orders = df[df['Sales'] > 3500]
print("\nOrders with Sales > 3500:")
print(high_sales_orders)

# Orders where Region = "East" and Sales > 2500
east_high_sales_orders = df[(df['Region'] == 'East') & (df['Sales'] > 2500)]
print("\nOrders in East region with Sales > 2500:")
print(east_high_sales_orders)
```

```python
# Q3 Customer Purchase Analysis

# Group by Customer and calculate total sales
customer_sales = df.groupby('Customer')['Sales'].sum()

# Sort in descending order and get top 5 customers
top_customers = customer_sales.sort_values(ascending=False).head(5)

print("Top 5 Customers based on Total Sales:")
print(top_customers)
```

```python
# Q4 Regional Sales Performance

# Calculate total sales by region
region_sales = df.groupby('Region')['Sales'].sum()

print("Total Sales by Region:")
print(region_sales)

# Identify region with highest revenue
top_region = region_sales.idxmax()
max_sales = region_sales.max()

print("\nRegion with Highest Revenue:")
print(top_region, "with sales =", max_sales)
```

```python
# Q5 Product Performance Analysis

# Calculate total quantity sold for each product
product_quantity = df.groupby('Product')['Quantity'].sum()

print("Total Quantity Sold for Each Product:")
print(product_quantity)

# Identify most popular product
top_product = product_quantity.idxmax()
max_quantity = product_quantity.max()

print("\nMost Popular Product:")
print(top_product, "with quantity =", max_quantity)
```

```python
# Q6 Multi-Level Grouping

# Group by Region and Category and calculate total sales
multi_group = df.groupby(['Region', 'Category'])['Sales'].sum()

print("Total Sales by Region and Category:")
print(multi_group)

# Optional: Convert to table format for better view
multi_group_table = multi_group.unstack()

print("\nFormatted Table:")
print(multi_group_table)
```

```python
# Q7 Pivot Table Creation

# Create pivot table
pivot_table = pd.pivot_table(df,values='Sales',index='Region',columns='Product',aggfunc='sum')

print("Pivot Table (Region vs Product - Total Sales):")
print(pivot_table)

# Find which region sells the most laptops
laptop_sales = pivot_table['Laptop']
top_region_laptop = laptop_sales.idxmax()
max_sales = laptop_sales.max()

print("\nRegion selling the most Laptops:")
print(top_region_laptop, "with sales =", max_sales)
```

```python
# Q8 Profitability Analysis

# Calculate average profit for each category
avg_profit = df.groupby('Category')['Profit'].mean()

print("Average Profit for Each Category:")
print(avg_profit)

# Identify most profitable category
top_category = avg_profit.idxmax()
max_profit = avg_profit.max()

print("\nMost Profitable Category:")
print(top_category, "with average profit =", max_profit)
```

```python
# Q9 Discount Impact Study

# Calculate average profit for each discount level
discount_profit = df.groupby('Discount')['Profit'].mean()

print("Average Profit for Each Discount Level:")
print(discount_profit)

# Sort for better understanding
sorted_discount_profit = discount_profit.sort_index()
print("\nSorted Discount vs Profit:")
print(sorted_discount_profit)
```

```python
# 10. Q10 End-to-End Data Wrangling Task

# Step 1: Filter dataset where Sales > 3000
filtered_df = df[df['Sales'] > 3000]

# Step 2: Group by Region and Product and calculate total sales
grouped_data = filtered_df.groupby(['Region', 'Product'])['Sales'].sum()

# Step 3: Convert into pivot table
pivot_result = grouped_data.unstack()

print("Final Pivot Table (Filtered Sales > 3000):")
print(pivot_result)
```
